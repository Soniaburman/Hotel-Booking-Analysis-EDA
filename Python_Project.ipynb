**Data Analysis and Visualizations**

```python
cancelled_percentage = df['is_canceled'].value_counts(normalize = True)
cancelled_percentage
```

```python
cancelled_percentage = df['is_canceled'].value_counts(normalize = True)
print(cancelled_percentage)

plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
```

Around 63% of bookings were not canceled, while about 37% were canceled, showing a significant cancellation rate. This indicates cancellations are a major factor affecting hotel bookings.

```python
plt.figure(figsize = (8,4))
ax = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
ax.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.show()
```

City Hotels have a higher number of total bookings compared to Resort Hotels. Cancellation rates are noticeably higher in City Hotels than in Resort Hotels, indicating different guest behaviors between hotel types.

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (18,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel') 
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

From 2016 to 2017, city hotel prices are generally lower than resort hotels, but they spike higher on weekends or holidays, likely due to increased demand during those times.

```python
df['month'] = df['reservation_status_date'].dt.month  
plt.figure(figsize = (12,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data =df, palette = 'bright') 
plt.title('Reservation status per month', size = 20)
plt.ylabel('number of reservations')
plt.legend(['not_cancelled','cancelled'])
plt.show()
```

The highest cancellations happened in January, while the lowest were in August. Interestingly, August also had the highest number of reservations, while December had the lowest. This suggests that when bookings are high, cancellations are low.

```python
cancelled_data = df[df['is_canceled'] == 1] 
top_10_country = cancelled_data['country'].value_counts()[:5] 
plt.figure(figsize = (6,6))
plt.title('Top 10 countries with reservation cancelled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index) 
plt.show()
```

Portugal (PRT) has the highest booking cancellations, around 70%. Hotels there could reduce cancellations by improving services, adjusting prices, offering discounts, and increasing promotions or advertising efforts.
